```python
plt.style.use('ggplot')
fig=plt.figure(figsize=(16,24))
#设置图像的大标题
fig.suptitle("未发送信息数据分析",fontsize=16,fontweight="bold")
print('标题设置成功！')
```

```python
import seaborn as sns
#添加第一个视图
ax1 = fig.add_subplot(4,2,1)
seaborn.boxplot(data = data,y = 'unsend_number')
plt.title('Box Plot of unsend messages')
plt.ylabel('total_unsend')
print('第一个视图完成！')

```

```python
#添加第二个视图
ax2 = fig.add_subplot(4,2,2)
sns.boxplot(data =data,x= 'qualified',y = 'unsend_number')
ax2.set(ylabel ='租车总人数', xlabel ='月份',title = '月份与租车人数箱形图')
print('第二个视图完成！')
plt.savefig('C:/Temp/是否具备资格1')
```

```python
#添加第二个视图
ax2 = fig.add_subplot(4,2,4)
sns.boxplot(data = data,x= 'client',y = 'unsend_number')
ax2.set(ylabel ='租车总人数', xlabel ='月份',title = '月份与租车人数箱形图')
print('第二个视图完成！')
```

```python
#添加第二个视图
ax2 = fig.add_subplot(4,2,3)
sns.boxplot(data = data,x= 'send_times',y = 'unsend_number')
ax2.set(ylabel ='租车总人数', xlabel ='月份',title = '月份与租车人数箱形图')
print('第二个视图完成！')
```
